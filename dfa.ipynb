```python
# Remove itens do Ste(com remove e discard)
# Com (remove) se o item não existir vai retornar um erro. Com o (discard) se o itens não existir vai retornar nenhum erro.

conjunto_de_nomes = {"Maria","João", "Jose", 'Jose'}

conjunto_de_nomes.remove ("Maria")

print(conjunto_de_nomes)

conjunto_de_nomes = {"Maria","João", "Jose", 'Jose'}

conjunto_de_nomes.discard("mariana")

print(conjunto_de_nomes)
```

```python
# Métodos úteis para Sets
# Além de remover dados duplicados, Set são úteis para comparar conjuntos(ver se tem diferença entre eles, o que tem igul, ver um conjunto está dentro do outro e tec)

#diferença(difference)

conjunto_de_nomes1 = {"Maria","João", "Jose"}
conjunto_de_nomes2 = {"Maria","João","Daniel"}

diferenca = conjunto_de_nomes1.difference(conjunto_de_nomes2)

print(diferenca)

```

```python
# intercecção o que eles tem igual:

conjunto_de_nomes1 = {"Maria","João", "Jose"}
conjunto_de_nomes2 = {"Maria","João","Daniel"}

igual = conjunto_de_nomes1.intersection(conjunto_de_nomes2)

print(igual)
```

```python
# Se um conjunto é subconjunto de outro maior:

conjunto_de_nomes_da_escolas = {"Maria", "Jose", "João"}
conjunto_de_nomes_da_aula = {"Maria", "Jose"}

se_subconjunto_eh_verdade = conjunto_de_nomes_da_aula.issubset(conjunto_de_nomes_da_escolas)

print(se_subconjunto_eh_verdade)
```

Exemplo prático, criando um mini chatbot organizando o nome e idade em dicionário.
verificandose tem mais de 18 anos (try & Except), e exportando dados em JSON

```python
continuar = "sim"
dicionario_de_pessoas = {}
indice = 1

while (continuar == "sim"):
  convidado = {}
  nome = input("Qual o nome do convidado? ")
  convidado["nome"] = nome
  idade = input("Qual a idade do convidado? ")
  convidado["idade"] = idade

  try:
    idade = int(idade)
    if(idade >=18):
      dicionario_de_pessoas[indice] = convidado
      indice = indice + 1
    else:
      print("O convidado precisar ter mais de 18 anos")

  except:
    print("Ocorreu um erro!")

  continuar = input("Alguém mais vem para festa? sim ou não: ")

print("Dicionário de Convidados/Pessoas")
arquivo_JSON = json.dumps(dicionario_de_pessoas, indent= 4)          

print(arquivo_JSON)
```

```python
x = df['fat']
y = df['calories']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(x, y)
plt.xlabel('Gordura')
plt.ylabel('calorias')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from numpy.core.fromnumeric import reshape
import numpy as np
from sklearn import linear_model

reg = linear_model.LinearRegression()

reg.fit(np.array(x_train).reshape(-1,1), np.array(y_train).reshape(-1,1))

reg.intercept_, reg.coef_
```

```python
# Previsão para treino ( como fizermos antes)
y_train_pred = reg.predict(np.array(x_train).reshape(-1,1))
y_train_pred.shape
```

```python
# Previsão para teste ( dados que não foram usados para treinar o modelo)
y_test_pred = reg.predict(np.array(x_test).reshape(-1,1))
y_test_pred.shape
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Qualidade no treino
print("---treino---")
print("RMSE:", mean_squared_error(y_train, y_train_pred, squared=False))
print("R²", r2_score(y_train, y_train_pred))
```

```python
# Qualidade de teste
print("---teste---")
print("RMSE:", mean_squared_error(y_test, y_test_pred, squared=False))
print("R²", r2_score(y_test, y_test_pred))
```

```python
import pandas as pd

url = 'https://vincentarelbundock.github.io/Rdatasets/csv/AER/GoldSilver.csv'

df = pd.read_csv(url, error_bad_lines=False)

df.head(10)
```

```python
x = df[['silver']]
y = df['gold']
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(x, y, 'og')
plt.xlabel('silver')
plt.ylabel('gold')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train, y_train)

reg.intercept_, reg.coef_
```

```python
y_train_pred = reg.predict(x_train)
y_test_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print("---Treino---")
print("RMSE:", mean_squared_error(y_train, y_train_pred, squared=False))
print("R²", r2_score(y_train, y_train_pred))
```

```python
print("---Teste---")
print("RMSE:", mean_squared_error(y_test, y_test_pred, squared=False))
print("R²", r2_score(y_test, y_test_pred))
```
